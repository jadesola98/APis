import necessary libraries

```python
import requests
import pandas as pd
import json
```

assign api to variable

```python
url = "http://api.tvmaze.com/search/shows?q=golden girls"
```

check error response for api

```python
data = requests.get(url).json()
```

```python
data
```

```python
data=pd.DataFrame(data)
data
```

```python
data.explode("show.g")
```

```python
pd.json_normalize(
    data, 
    record_path =["categories","_links.curies"], 
    meta=['teleport_city_score'],
    errors='ignore'
)
```

```python
d = data.explode("categories", ignore_index=True)
d = d.explode("_links.curies", ignore_index=True)
d
```

```python
d=pd.json_normalize(json.loads(d.to_json(orient="records")))
d
```
