import necessary libraries

```python
import requests
import pandas as pd
```

assign api to variable

```python
url = "http://universities.hipolabs.com/search?country=United+Kingdom"
```

check error response for api

```python
requests.get(url)
```

inspect data

```python
requests.get(url).json()
```

assign result to a variable

```python
data = requests.get(url).json()
```

flatten JSON into DataFrame

```python
pd.json_normalize(data)
```

```python
pd.json_normalize(data).columns
```

using the argument record_path to flatten the JSON further

```python
pd.json_normalize(
    data, 
    record_path =['domains'], 
    meta=['web_pages', 'state-province', 'name', 'country',
       'alpha_two_code']
)
```
